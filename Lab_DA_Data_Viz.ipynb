```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('/content/sample_placement.csv') 
df.head(15) 
```

```python
plt.figure(figsize=(16,8))
sns.boxplot(data=df) 
plt.show() 
```

```python
plt.figure(figsize=(16,8))
sns.histplot(data=df['IITM'], bins=80, kde=True) 
plt.title('IIT Madras Placements') 
plt.show() 
```

```python
plt.figure(figsize= (16,8))
sns.histplot(df['VITV'], bins=80, kde=True)
plt.title('VIT Vellore Placements') 
plt.show() 
```

```python
plt.figure(figsize=(16,8)) 
sns.histplot(df['IITK'], bins=80, kde=True) 
plt.title('IIT Kharagpur Placements') 
plt.show() 
```

```python
plt.figure(figsize=(16,8))
sns.histplot(df['BITSP'], bins=80, kde=True)
plt.title('BITS Pilani Placements') 
plt.show() 
```

```python
plt.figure(figsize=(16,8)) 
sns.histplot(df['IIMT'], bins=80, kde=True) 
plt.title('IIM Trichy Placements') 
plt.show() 
```

```python
plt.figure(figsize=(16,8))
sns.histplot(data=df, bins=80, kde=True) 
plt.show() 
```
